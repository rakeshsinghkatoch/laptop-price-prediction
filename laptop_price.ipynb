## converting data into numericals

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder=preprocessing.LabelEncoder()

df['Company']=label_encoder.fit_transform(df['Company'])
df['Product']=label_encoder.fit_transform(df['Product'])
df['TypeName']=label_encoder.fit_transform(df['TypeName'])
df['ScreenResolution']=label_encoder.fit_transform(df['ScreenResolution'])
df['CPU_Company']=label_encoder.fit_transform(df['CPU_Company'])
df['CPU_Type']=label_encoder.fit_transform(df['CPU_Type'])
df['Memory']=label_encoder.fit_transform(df['Memory'])
df['GPU_Company']=label_encoder.fit_transform(df['GPU_Company'])
df['GPU_Type']=label_encoder.fit_transform(df['GPU_Type'])
df['OpSys']=label_encoder.fit_transform(df['OpSys'])

```

```python
df.dtypes
```

## Data scaling

```python
def feature_scale(scale):
    
    numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
    new_df = df.select_dtypes(include=numerics)
    if scale == 'minmax':
        
        scaler = MinMaxScaler()
    else:
        scaler = StandardScaler()
    df_scaled = pd.DataFrame(scaler.fit_transform(new_df.to_numpy()),columns=new_df.columns)
    return df_scaled
```

```python
scale = 'minmax' #standard
df_scaled=feature_scale(scale)
df_scaled
```

```python
corr=df_scaled.corr()
corr.style.background_gradient(cmap='coolwarm')
```

### All the attributes are corelated to the target value so keeping all the attributes

```python
features=df_scaled.columns[:-1]
features          
```

```python
label=df_scaled.columns[-1]
label
```

```python
X=df_scaled[features]
y=df_scaled[label]
```

```python
X
```

```python
y
```

## Spliting data into test and train 75:25 ratio

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

## creating linear regressin model

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
s1=model.fit(X_train,y_train)
```

```python
s_r_sq=s1.score(X_train, y_train)
```

```python
print('coefficient of determination:', s_r_sq)
print('intercept:', s1.intercept_)
print('slope:', s1.coef_)
```

## Prediction

```python
y_pred = s1.predict(X_test)
print('predicted response:', y_pred, sep='\n')
```

## Evaluating the model

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
print("R-Squared on train dataset={}".format(s1.score(X_test,y_test)))
s1.fit(X_test,y_test)
print("R-Squaredon test dataset={}".format(s1.score(X_test,y_test)))
```
